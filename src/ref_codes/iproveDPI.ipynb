```python
import cv2
import os
import numpy as np
from pathlib import Path
import matplotlib.pyplot as plt

folder = 'resultCascadeSpecial'
pathToSave = 'resultTesseract'

# dir = os.listdir(folder)
# for filename in dir: #os.listdir(folder):
#     img = cv2.imread(os.path.join(folder,filename))
#     plt.imshow(img[:,:,::-1])
#     plt.show()
```

```python
sr = cv2.dnn_superres.DnnSuperResImpl_create()


# path = "EDSR_x4.pb"
# sr.readModel(path)
# sr.setModel("edsr",4)

path = "ESPCN_x3.pb"
sr.readModel(path)
sr.setModel("espcn",3)


dir = os.listdir(folder)
for filename in dir:
    img = cv2.imread(os.path.join(folder,filename))
    result = sr.upsample(img.copy())
    resized = cv2.resize(img,dsize=None,fx=4,fy=4)
    cv2.imwrite(os.path.join(pathToSave , filename), result)
    plt.subplot(1,2,1)
    plt.imshow(result[:,:,::-1])
    plt.subplot(1,2,2)
    plt.imshow(resized[:,:,::-1])
    plt.show()
```
